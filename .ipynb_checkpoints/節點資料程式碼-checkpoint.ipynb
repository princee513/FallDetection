```python
import cv2
import numpy as np
import os
import tensorflow as tf 
import matplotlib.pyplot as plt
import pickle
```

```python
EDGES = {
    (0, 1): 'm',
    (0, 2): 'c',
    (1, 3): 'm',
    (2, 4): 'c',
    (0, 5): 'm',
    (0, 6): 'c',
    (5, 7): 'm',
    (7, 9): 'm',
    (6, 8): 'c',
    (8, 10): 'c',
    (5, 6): 'y',
    (5, 11): 'm',
    (6, 12): 'c',
    (11, 12): 'y',
    (11, 13): 'm',
    (13, 15): 'm',
    (12, 14): 'c',
    (14, 16): 'c'
}
```

```python
#3. Draw Keypoints
def draw_keypoints(frame, keypoints, confidence_threshold):
    y, x, c = frame.shape
    shaped = np.squeeze(np.multiply(keypoints, [y,x,1]))
    
    for kp in shaped:
        ky, kx, kp_conf = kp
        if kp_conf > confidence_threshold:
            cv2.circle(frame, (int(kx), int(ky)), 4, (0,255,0), -1) 
            
def draw_connections(frame, keypoints, edges, confidence_threshold):
    y, x, c = frame.shape
    shaped = np.squeeze(np.multiply(keypoints, [y,x,1]))
    
    for edge, color in edges.items():
        p1, p2 = edge
        y1, x1, c1 = shaped[p1]
        y2, x2, c2 = shaped[p2]
        
        if (c1 > confidence_threshold) & (c2 > confidence_threshold):      
            cv2.line(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0,0,255), 2)
            
```

```python
# 定义摔倒检测函数
image_folder = r'C:\Users\B00\Desktop\bodyImage\fornobody'
model_path = r'C:\Users\B00\Desktop\for project\movenet file\lite-model_movenet_singlepose_lightning_3.tflite'
confidence_threshold = 0.2


interpreter = tf.lite.Interpreter(model_path=r'C:\Users\B00\Desktop\for project\movenet file\lite-model_movenet_singlepose_lightning_3.tflite')
interpreter.allocate_tensors()

# 初始化一个字典来存储关键点数据
keypoints_dict = {}

for filename in os.listdir(image_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        # 读取图像
        image_path = os.path.join(image_folder, filename)
        image = cv2.imread(image_path)
        
        # Reshape image
        input_image = tf.cast(image, dtype=tf.float32)

        # Setup input and output
        input_details = interpreter.get_input_details()
        output_details = interpreter.get_output_details()

        # Make predictions
        input_image = np.expand_dims(input_image, axis=0)  # 添加批次维度
        interpreter.set_tensor(input_details[0]["index"], input_image)
        interpreter.invoke()
        keypoints_with_scores = interpreter.get_tensor(output_details[0]["index"])
        print(keypoints_with_scores)
        
        # 检测摔倒
       

        # 在这里，我们将关键点数据添加到字典中
        keypoints_dict[filename] = keypoints_with_scores
        
        # Rendering 
        draw_connections(image, keypoints_with_scores, EDGES, 0.4)
        draw_keypoints(image, keypoints_with_scores, 0.4)
    
        # 显示图像
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()

# 在循环外部，保存字典到文件中
with open('keypoints_dict.pkl', 'wb') as f:
    pickle.dump(keypoints_dict, f)
```
