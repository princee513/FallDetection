```python
import cv2
import numpy as np
from keras.models import load_model
```

```python
# 載入模型
model = load_model(r"D:\falldetection\trainedModelforCNN\theFirst")

#讀取照片
cap = cv2.imread('IU1.jpg')

image_height = 192
image_width  = 192

while(cap.isOpened()):
    # 讀取一幀
    ret, frame = cap.read()
    if ret == True:
        # 調整幀的大小並正規化
        frame = cv2.resize(frame, (image_height, image_width))
        frame = frame / 255

        # 添加一個維度以匹配模型的輸入形狀
        frame = np.expand_dims(frame, axis=0)

        # 使用模型進行預測
        prediction = model.predict(frame)

        # 找出最可能的類別
        class_index = np.argmax(prediction)

        # 做出相應的處理，例如如果class_index表示"跌倒"，則發出警告
        if class_index == 0:
            print("警告：偵測到站著！")
        elif class_index == 1:
            print("警告：偵測到坐著！")
        elif class_index == 2:
            print("警告：偵測到躺著！")
            break
    else:
        break

# 釋放資源
cap.release()
cv2.destroyAllWindows()
```
